```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
# Điền giá trị thiếu bằng giá trị trung bình 
df['Income']=df['Income'].fillna(df['Income'].mean())
```

```python
df
```

```python
#
# Rời rạc hóa giá trị thuộc tính Age, Credit Rating bằng equal-frequency: chia thành k bins, mỗi bin có số mẫu xấp xỉ nhau
# Age:
# Bin 1: 21, 29, 30, 30, 32
# Bin 2: 38, 39, 45, 45, 46
# Bin 3: 47, 48, 50, 51
#
discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
df[['Age']] = discretizer.fit_transform(df[['Age']])
```

```python
#
# Rời rạc hóa giá trị thuộc tính Income, Credit Rating bằng equal-width: chia thành k bins, mỗi bin có 'width' = (max - min) / k
# Income: width = (80000 - 25000)/3 = 18333.3
# Bin 1: Income >= 25000 & Income <= 43333.3
# Bin 2: Income > 43333.3 & Income <= 61666.3
# Bin 3: Income > 61666.3 & Income <= 80000
#
discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
df[['Income']] = discretizer.fit_transform(df[['Income']])
#
# Credit Rating:
# Bin 1: 2, 2, 2, 2, 3, 3, 3, 3 (>= 2 & < 4)
# Bin 2: 4, 4, 5, 5, 6, 6 (>= 4 & <= 6)
#
discretizer = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='uniform')
df[['Credit Rating']] = discretizer.fit_transform(df[['Credit Rating']])
```

```python
df.to_excel("output.xlsx")  
```
