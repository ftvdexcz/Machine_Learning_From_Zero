```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, KBinsDiscretizer, LabelEncoder
from sklearn.linear_model import LinearRegression
```

**1.Hiển thị dữ liệu**

```python
df = pd.read_excel('BTH2.xlsx')
df
```

**2.Điền giá trị thiếu theo phương pháp trung bình**

```python
# Age
print(df['Age'].mean())
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
# Income
print(df['Income'].mean())
df['Income']=df['Income'].fillna(df['Income'].mean())
```

```python
df[['Online Shopper']] = df[['Online Shopper']].apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
df2 = df.copy()
df
```

**3.Các bản ghi có Income>70000 & Age>50**

```python
df.loc[(df['Income']>70000) & (df['Age']>50)]
```

**4.Chuẩn hóa dữ liệu theo min max**

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income']])
print(scaler.data_max_, scaler.data_min_)
```

```python
df[['Income']] = scaler.transform(df[['Income']])
df[['Age']] = scaler.fit_transform(df[['Age']])
```

```python
df
```

**5.Liên tục hóa thuộc tính Region**

```python
dummies = pd.get_dummies(df['Region'])
df = pd.concat([df, dummies], axis=1)
df
```

**6.Xây dựng mô hình dự đoán cho thuộc tính Online Shopper**

```python
X = df.drop(['Region','Online Shopper'], axis=1).values
print(X)
```

```python
le = LabelEncoder()
y = df['Online Shopper'].values
y = le.fit_transform(y)
print(y)
```

```python
clf = LinearRegression().fit(X,y)
```

```python
print(f'Accuracy score: {clf.score(X, y)*100} %')
```

**8.Rời rạc hóa thuộc tính Income**

```python
df2
```

```python
# equal-width: chia thành k bins, mỗi bin có 'width' = (max - min) / k
discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
df2[['Income']] = discretizer.fit_transform(df2[['Income']])
```

```python
dic = {
  0.: 'Low',
  1.: 'Medium',
  2.: 'High'
}

df2['Income'] = [dic[x] for x in df2['Income'].values]
```
